```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
# Getting labels
with open('qsarvars.txt', 'r') as outfile:
    data = outfile.readlines()

    labels = []

    for line in data:
        if 'NAME' in line:
            label = line
            label = label.split()
            labels = label[2:]
            labels = labels[:-2]

    descriptors = np.genfromtxt(data, delimiter='\t')
    descriptors = descriptors[1:, 2:]
```

```python
# Calculating correlation
def cor_coefficient(x, y):
    cor_matrix = np.corrcoef(x, y)
    return cor_matrix[0, 1]

dim = descriptors.shape
correlations = np.zeros((dim[1] - 2, dim[1] - 2))

for n in range(dim[1] - 2):
    for m in range(dim[1] - 2):
        x_var = descriptors[:, m]
        y_var = descriptors[:, n]
        correlations[m, n] = cor_coefficient(x_var, y_var)
print(correlations)
```

```python
desc_intercor = pd.DataFrame(index=labels, columns=labels, data=correlations)
```

```python
mask = np.triu(np.ones_like(desc_intercor, dtype=bool))
mask = mask[1:, :-1]
corr = desc_intercor.iloc[1:, :-1].copy()

fig, ax = plt.subplots(figsize=(8, 8), dpi=2400)
sb.heatmap(corr, annot=False, linewidths=0, mask=mask,
           vmin=-1, vmax=1, xticklabels=True,
           yticklabels=True, cbar_kws={'shrink': 0.5},
           cmap='viridis')
sb.color_palette('viridis', as_cmap=True)
plt.xticks(fontsize=3)
plt.yticks(fontsize=3)

plt.title('Descriptor Intercorrelation')
plt.savefig('desc_intercor1.png', dpi=2400)
```
