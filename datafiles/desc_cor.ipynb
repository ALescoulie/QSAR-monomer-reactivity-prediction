```python
import pandas as pd
import numpy as np
import seaborn
```

```python
# Import data
full_data = pd.read_table('qsarvars.txt', sep='\t')

# Getting U and V values
u_values = full_data['U'].tolist()
v_values = full_data['V'].tolist()
u_cor = []
v_cor = []

des_names = full_data.columns.values.tolist()
des_names = des_names[2:125]
```

```python
# Calculating correlations
for col in des_names:
    des_values = full_data.get(col)
    u = np.corrcoef(u_values, des_values)
    v = np.corrcoef(v_values, des_values)
    u_cor.append(u[0, 1])
    v_cor.append(v[0, 1])

```

```python
# Splitting up data for manageability of figure sizing

des_cor0 = pd.DataFrame({'U': u_cor[0:19],
                        'V': v_cor[0:19]},
                        index=des_names[0:19])

des_cor1 = pd.DataFrame({'U': u_cor[20:39],
                         'V': v_cor[20:39]},
                        index=des_names[20:39])

des_cor2 = pd.DataFrame({'U': u_cor[40:59],
                         'V': v_cor[40:59]},
                        index=des_names[40:59])

des_cor3 = pd.DataFrame({'U': u_cor[60:79],
                         'V': v_cor[60:79]},
                        index=des_names[60:79])

des_cor4 = pd.DataFrame({'U': u_cor[80:99],
                         'V': v_cor[80:99]},
                        index=des_names[80:99])

des_cor5 = pd.DataFrame({'U': u_cor[100:119],
                         'V': v_cor[100:119]},
                        index=des_names[100:119])

des_cor6 = pd.DataFrame({'U': u_cor[119:],
                         'V': v_cor[119:]},
                        index=des_names[119:])
```

```python
import matplotlib.pyplot as plt
fig, (ax0, ax1, ax2, ax3, ax4, ax5) = plt.subplots(ncols=6, figsize=(10, 6))
fig.subplots_adjust(wspace=1.25)

seaborn.heatmap(des_cor0, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax0)
seaborn.heatmap(des_cor1, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'viridis': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax1)
seaborn.heatmap(des_cor2, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax2)
seaborn.heatmap(des_cor3, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax3)
seaborn.heatmap(des_cor4, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax4)
seaborn.heatmap(des_cor5, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax5)
plt.savefig('descriptors.png', dpi=1200)
```
